## Autocorrelation 

```python
# The eigenvals
region = 'NH'
kas = [0.25,0.5,1,2.5,5,10,20,40,80,160,320]
for eof in [0,1]:
    datas = []
    for ka in kas:
    #     data = xr.open_dataset(f'/birner-home/ldavis/data/hs{ka:07.3f}/4xdaily_inst.pcs{region}.nc', 
    #                            decode_times=False)['PC1'].isel(pfull=-1) # last one is almost 850mb
    #     datas += [np.var(data.values.squeeze())]
        data = xr.open_dataset(f'/birner-home/ldavis/data/hs{ka:07.3f}/4xdaily_inst.evals{region}.nc', 
                               decode_times=False)['EOF'].isel(pfull=-1).isel(time=eof)
        datas += [data.values.squeeze()]
    py.setup(size2=10)
    f, ax = py.subplots(width=3, hspace=0, top=.5, wspace=0, aspect=1)
    ax.plot(kas, datas, color='k', lw=1.5, 
          clip_on=False)
    if eof==0:
        ystuff = {'ylim':(0,10), 'ylocator':2}
    else:
        ystuff = {'ylim':(0,3.1), 'ylocator':.5}
    ax.format(xlim=[0.25,320], xscale='log', xlocator=[0.1,.5,1,5,10,50,100], 
            xgrid=True, ygrid=True, xformatter=py.Formatter(),
            yscale='linear', xlabel='radiative timescale (days)', ylabel='PC variance',
            title=f'EOF{eof+1:d} eigenvalue: NH 850mb $\\overline{{u}}$', **ystuff)
    f.save(f'/birner-home/ldavis/0_evals-all.pdf')
    # The autocorrelation
    for lag in [3*4, 10*4, 20*4]:
        datas2 = []
        for ka in kas:
        #     lag = 10*4 # so 10 days, 40 timesteps?
            data2 = xr.open_dataset(f'/birner-home/ldavis/data/hs{ka:07.3f}/4xdaily_inst.pcs{region}.nc', 
                                   decode_times=False)[f'PC{eof+1:d}'].isel(pfull=-1) # last one is almost 850mb
            datas2 += [np.corrcoef(data2.values.squeeze()[lag:], data2.values.squeeze()[:-lag])[0,1]]
        py.setup(size2=10)
        f, ax = py.subplots(width=3, hspace=0, top=.5, wspace=0, aspect=1)
        ax.plot(kas, datas2, color='k', lw=1.5, 
              clip_on=False)
        ax.format(xlim=[0.25,320], xscale='log', xlocator=[0.1,.5,1,5,10,50,100], 
                xgrid=True, ygrid=True, xformatter=py.Formatter(),
                yscale='linear', xlabel='radiative timescale (days)', ylabel='PC variance',
                 ylim=(-.2,1), ylocator=.2, title=f'Lag-{lag//4:d} EOF{eof+1:d} $\\rho$: NH 850mb $\\overline{{u}}$')
        f.save(f'/birner-home/ldavis/0_autocorrs-all.pdf')
    # The e-folding timescale
    for lag in [3*4, 10*4, 20*4]:
        datas2 = []
        for ka in kas:
        #     lag = 10*4 # so 10 days, 40 timesteps?
            data2 = xr.open_dataset(f'/birner-home/ldavis/data/hs{ka:07.3f}/4xdaily_inst.pcs{region}.nc', 
                                   decode_times=False)[f'PC{eof+1:d}'].isel(pfull=-1) # last one is almost 850mb
            datas2 += [np.corrcoef(data2.values.squeeze()[lag:], data2.values.squeeze()[:-lag])[0,1]]
        py.setup(size2=10)
        f, ax = py.subplots(width=3, hspace=0, top=.5, wspace=0, aspect=1)
        ax.plot(kas, datas2, color='k', lw=1.5, 
              clip_on=False)
        ax.format(xlim=[0.25,320], xscale='log', xlocator=[0.1,.5,1,5,10,50,100], 
                xgrid=True, ygrid=True, xformatter=py.Formatter(),
                yscale='linear', xlabel='radiative timescale (days)', ylabel='PC variance',
                 ylim=(-.2,1), ylocator=.2, title=f'Lag-{lag//4:d} EOF{eof+1:d} $\\rho$: NH 850mb $\\overline{{u}}$')
        f.save(f'/birner-home/ldavis/0_autocorrs-all.pdf')
```

```python
# Draw plot of different EOFS for each model level
cmap = 'viridis'
legend = False # colorbar instead
levidxs = None # range of level indices
cmap = plt.cm.get_cmap(cmap)
nplot = 2 # number of EOFS
#def pressure(levidx):
#    return -12.5+1000*(levidx/40) # approx pressure level of bottom boundary
# Iterate through EOFS from different latitude regions
dfs = {} # will store percent variance explained
regions = ['NH','SH']
for region in regions:
    f, ax = py.subplots(nrows=nplot,ncols=1,sharex=True,spany=False,
            width=5,hspace=0.2,bottom=.4,legend_width=.8,
            bottomcolorbar=True)
    data = xr.open_dataset(f'4xdaily_inst.eofs{region}.nc', decode_times=False)
    eig = xr.open_dataset(f'4xdaily_inst.evals{region}.nc', decode_times=False)
    for a,ieof in zip(ax,range(nplot)):
        hs = []
        fracs = [] # fractions
        a.axhline(0, dashes=(5,5), color='k', linewidth=py._Settings().spine['linewidth'])
        for levidx in range(data.dims['pfull']):
            # Data (and change latitudes if in SH)
            name = f'PC{ieof+1:d}'
            color = cmap(levidx/data.dims['pfull'])
            eof, lat = data[name].isel(pfull=levidx).values.squeeze(), data['lat'].values
            if region=='SH':
                lat, eof = -1*np.flipud(lat), np.flipud(eof)
            # Scale, and try to match sign conventions
            eof *= np.sign(np.mean(eof[0:20]))
            # Get fractional variance (and take average across levels to display)
            name = ('EOF' if 'EOF' in eig else 'EOF1')
            fracs.append(eig[name].isel(pfull=levidx).isel(time=ieof).values.squeeze()\
                          /eig[name].isel(pfull=levidx).values.sum())
            # Plot
            #p = pressure(levidx)
            #ps.append(pfull)
            hs.extend(a.plot(lat, eof, zorder=2, color=color, linewidth=1, linestyle='-', dashes=(), 
                      label=fr'{p:.0f}mb'))
        # Print nice summary, to see which levels have more variance explained EOF1
        if ieof+1 not in dfs:
            dfs[ieof+1] = []
        dfs[ieof+1].append(pd.Series(fracs,index=data.coords['pfull'].values))
        #print(f'EOF{ieof+1:d}: ' + ','.join(f'{p:.0f}mb: {100*f:.1f}%' for p,f in zip(ps,fracs)))
        # Format
        title = ('' if ieof>0 else f'{region if len(region)==2 else region.title()}: ' +
                 f'EOFS 1-{nplot:d} of control Held-Suarez run zonal wind')
        xlim = ((-90,90) if region=='GLOBE' else (0,90))
        a.format(xlim=xlim, ylim=(-7,7), yminorlocator=1, ylocator=2, xminorlocator=5,
                titleinside=f'{100*min(fracs):.1f}%-{100*max(fracs):.1f}% variance',
                ygrid=True, xgrid=True,
                xlabel='latitude', ylabel=f'EOF {ieof+1:d} (m/s)',
                title=title)
    # And make colorbar (create mappable object first)
    pedges = np.linspace(data.coords['pfull'].min()-12.5,
                         data.coords['pfull'].max()+12.5, data.dims['pfull']+3) # why +3? dunno, but it works
    m = a.contourf(np.array([0,0]),np.array([0,0]),np.zeros([2,2]), 
                    vmin=data.coords['pfull'].min(), vmax=data.coords['pfull'].max(),
                    levels=pedges)
    if legend:
        f.bottomlegend.format(handles=hs, legend_kw={'ncol':5,'frameon':False,
                                                 'handletextpad':.3,
                                                 'columnspacing':1,'loc':'upper center'})
    else:
        f.bottomcolorbar.format(mappable=m, clabel='pressure (mb)', 
                                cgrid=False, clocator=np.arange(0,1000,100))
    # Finally save figure
    f.savefig(f'eof{region}.pdf', format='pdf', bbox_inches='tight')
for eof in dfs:
    dfs[eof] = pd.concat(dfs[eof],keys=regions,axis=1)
```

```python
# Also plot variance explained
hs = []
colors = ['C0','C1']
f, ax = py.subplots(width=5, bottomlegend=True, bottom=.4)
for eof,dftotal in dfs.items():
    for name,color in zip(dftotal.columns,colors):
        if 'globe' in name.lower(): continue
        #if eof==1: continue
        #if eof>1: continue
        x = dftotal[name].values/max(np.nanmax(dftotal[n].values) for n in dftotal.columns)
        y = dftotal.index.values
        filt = np.isfinite(x)
        hs.extend(ax.plot(x[filt], y[filt], color=color, 
                          dashes=(() if eof==1 else (1,1)),
                          label=f'{name} EOF{eof:d}'))
ax.format(ylim=(900,300), ylocator=np.arange(0,1000,100), yminorlocator=np.arange(0,1000,50), 
        xgrid=True, ygrid=True, xlocator=np.arange(0,2,.1), xminorlocator=.05,
        xlabel='variance explained relative to max of all levels in each hemisphere', ylabel='pressure (mb)',
        xlim=(.7,1.05), title='Relative variance explained by EOFs at each level')
py.setup(legend_kw={'ncol':4, 'frameon':False})
f.bottomlegend.format(handles=hs)
f.savefig('evals.pdf')
```

```python
# Draw plot of different EOFS for each model level
cmap = 'viridis'
legend = False # colorbar instead
levidxs = [13, 36] # range of level indices
cmap = plt.cm.get_cmap(cmap)
regions = ['NH','SH']
single = False
nplot = 2 # number of EOFS
# Iterate through EOFS from different latitude regions
for single,pc,skip in zip([True,False,False],[1,1,2],[5,20,20]):
    name = f'PC{pc:d}'
    f, ax = py.subplots(nrows=len(regions),ncols=1,sharex=True,spany=False,
        aspect=1.8,width=6,hspace=0.2,bottom=.4,legend_width=.8,
        bottomcolorbar=(not legend and not single), bottomlegend=(legend and not single))
    for i,(a,region) in enumerate(zip(ax,regions)):
        hs = []
        data = xr.open_dataset(f'4xdaily_inst.pcs{region}.nc', decode_times=False)
        a.axhline(0, dashes=(5,5), color='k', linewidth=py._Settings().spine['linewidth'])
        pick = np.argmin(np.abs(data.coords['pfull'].values-687.5))
        for levidx in range(data.dims['pfull']):
            # Data (and change latitudes if in SH)
            if levidx!=pick and single:
                continue
            ndaily = 4 # number of recordings per day
            color = cmap(levidx/data.dims['pfull'])
            ts, time = data[name].isel(pfull=levidx).values.squeeze(), data['time'].values
            # Scale, and try to match sign conventions
            ts *= np.sign(np.mean(ts[:200]))
            # Plot
            hs.extend(a.plot(time[::skip*ndaily], ts[::skip*ndaily], 
                      color=('k' if single else color),
                      zorder=2, linewidth=1.5, linestyle='-', dashes=(), 
                      label=fr'{p:.0f}mb'))
            if single:
                a.scatter(time[::100*ndaily], ts[::100*ndaily],
                      color='r', zorder=3,
                      marker='o', s=5)
        # Add vertical lines at restart file locations
        choices=((3300,5700) if region=='SH' else (2300,3700))
        for x in range(1000,6100,100):
            a.axvline(x, color=('red' if x in choices and pc==1 else 'gray'), linewidth=1, 
                      alpha=(1 if x in choices and pc==1 else .25))
        # Format
        title = ('' if i>0 else f'{name} of control Held-Suarez run zonal wind {"at 700mb" if single else ""}')
        a.format(xlim=(1000,6000), ylim=(-3,3), yminorlocator=0.5, ylocator=1, xminorlocator=100, xlocator=500,
                titleinside=(region if len(region)==2 else region.title()),
                ygrid=False, xgrid=False,
                xlabel='day', ylabel=f'PC strength (stdevs)',
                title=title)
    # And make colorbar (create mappable object first)
    pedges = np.linspace(data.coords['pfull'].min()-12.5,
                         data.coords['pfull'].max()+12.5, data.dims['pfull']+1) # why +3? dunno, but it works
    m = a.contourf(np.array([0,0]),np.array([0,0]),np.zeros([2,2]), 
                    vmin=data.coords['pfull'].min(), vmax=data.coords['pfull'].max(),
                    levels=pedges)
    if legend and not single:
        py.setup(legend_kw={'ncol':5,'frameon':False,
                                    'handletextpad':.3,
                                    'columnspacing':1,'loc':'upper center'})
        f.bottomlegend.format(handles=hs,)
    if not legend and not single:
        f.bottomcolorbar.format(mappable=m, clabel='pressure (mb)', 
                                clocator=np.arange(0,1000,100), cgrid=False)
    # Finally save figure
    f.savefig(f'pc{pc:d}{"_1lev" if single else ""}.pdf', format='pdf', bbox_inches='tight')
```
